# 1. Mass vs Metabolism
(adapted from [Data Carpentry for Biologists](https://datacarpentry.org/semester-biology/))

The relationship between the body size of an organism and its metabolic rate is one of the most well studied and still most controversial areas of organismal physiology. We want to graph this relationship in the [Artiodactyl](https://en.wikipedia.org/wiki/Even-toed_ungulate) using a subset of data from a large compilation of body size data (Savage et al. 2004):

````python
body_mass = [32000, 37800, 347000, 4200, 196500, 100000, 4290, 
32000, 65000, 69125, 9600, 133300, 150000, 407000, 115000, 67000, 
325000, 21500, 58588, 65320, 85000, 135000, 20500, 1613, 1618]

metabolic_rate = [49.984, 51.981, 306.770, 10.075, 230.073, 
148.949, 11.966, 46.414, 123.287, 106.663, 20.619, 180.150, 
200.830, 224.779, 148.940, 112.430, 286.847, 46.347, 142.863, 
106.670, 119.660, 104.150, 33.165, 4.900, 4.865]
````

### Exercises

- 1.1 Create a `DataFrame` from these two lists and make a plot for body mass vs. metabolic rate.
- 1.2 Create a second plot for log(body mass) vs. log(metabolic rate). To set a log scale for both axes use:

```
ax = plt.gca()
ax.set(xscale="log", yscale="log")
```

> Think about what the shape of these graphs tells you about the relationship between mass and metabolic rate.

````python
### 1.1
# create pandas data frame from the two lists and name the columns according to the variable
print("\n 1.1")

df = pd.DataFrame(
    {'body_mass': body_mass,
     'metabolic_rate': metabolic_rate,
    })

# plot the data points using a regression line to show the relationship between the variables
sns.regplot(
    data = df,
    x="body_mass", 
    y="metabolic_rate"
)
plt.title("Body Mass vs. Metabolic Rate")
plt.xlabel("Body Mass")
plt.ylabel("Metabolic Rate")
plt.show()

### 1.2
print("\n 1.2")

# plot log values
sns.regplot(
    data = df,
    x="body_mass", 
    y="metabolic_rate"
)
# set a log scale for both axes
ax = plt.gca()
ax.set(xscale="log", yscale="log")
plt.title("Log Plot: Body Mass vs. Metabolic Rate")
plt.xlabel("Log Body Mass")
plt.ylabel("Log Metabolic Rate")

plt.show()
````

# 4. Mammal Body Size
(adapted from [Data Carpentry for Biologists](https://datacarpentry.org/semester-biology/))

There were a relatively large number of extinctions of mammalian species roughly 10,000 years ago. To help understand why these extinctions happened scientists are interested in understanding whether there were differences in the body size of those species that went extinct and those that did not.

The dataset `'data/mammal.tsv'` is the largest dataset on mammalian body size in the world. This dataset has data on the mass of recently extinct mammals as well as extant mammals (i.e., those that are still alive today). Take a look at the [original dataset's metadata](https://web.archive.org/web/20191219110715/http://www.esapubs.org/archive/ecol/E084/094/metadata.htm) to understand the structure of the data. One key thing to remember is that species can occur on more than one continent, and if they do then they will occur more than once in this dataset. 

The data also contains entries marked as *historical*, those can be considered *extinct*. It also containse entries marked *introduction*, those can be considered *extant*. You can use the pandas method `replace()` to change those values in your `DataFrame`.

Lets explore the data.

### Exercises

- 4.1 Determine the number of species. Remember that a species is uniquely defined by the combination of its genus name and its species name. Print the result to the screen. The number should be somewhere between 4000 and 5000.
- 4.2 Find out how many of the species are extinct and how many are extant, print the result to the screen. **HINT:** first separate the data into the extinct and extant components and then count the number of species.
- 4.3 Find out how many families are present in the dataset.
- 4.4 Now print the genus name, the species name, and the mass of the largest and smallest species. Note that it is not possible for a mammal to have negative mass. 
- 4.5 Create a plot that shows the average body size for extinct species versus extant species. Make sure that is directly clear from the plot which is which. What kind of plot should you use here?


````python
# add names to columns
col_names = ['continent', 'status', 'order', 'family', 'genus', 'species', 'log_mass', 'combined_mass', 'reference']
data_mammal = pd.read_csv('data/mammal.tsv', sep = '\t', na_values=['-999', '-999.00'], names = col_names, keep_default_na=False) 
# display(data_mammal.head(5))

# replace status values
data_mammal['status'] = data_mammal['status'].replace({
    'historical': 'extinct',
    'introduction': 'extant'
})

### 4.1 
# select the genus and species columns
genus_species = data_mammal[['genus', 'species']]

# remove duplicates to get unique species
unique = genus_species.drop_duplicates()

# count the rows in unique species
n_species = len(unique)

print("\n4.1 \nNumber of unique species:", n_species)

### 4.2 
print("\n4.2")

# select the status for extinct and extant separately and count the number of rows for each
extant = data_mammal['status'] == 'extant'
n_extant = len(data_mammal[extant])
print("Number of extant species:", n_extant)

extinct = data_mammal['status'] == 'extinct'
n_extinct = len(data_mammal[extinct])
print("Number of extinct species:", n_extinct)

### 4.3 
print("\n4.3")

# select family column and count the number of unique values in it
n_families = data_mammal['family'].nunique()
print("Number of families:", n_families)

### 4.4 
print("\n4.4")

# function for converting grams to kilograms 
def convert_g(grams):
    return grams * 0.0001

# add column for mass in kilograms for readability
data_mammal['combined_mass_kg'] = data_mammal['combined_mass'].map(convert_g)

# get the mass and ID of the largest mammal and use the ID to locate the genus and species name in other columns
largest_id = data_mammal['combined_mass'].idxmax()
largest_mass = data_mammal['combined_mass_kg'].max()
largest_genus = data_mammal['genus'][largest_id]
largest_species = data_mammal['species'][largest_id]
print(f"Largest species: {largest_species}, in genus: {largest_genus}, with mass: {largest_mass} kg")

# the smallest mammal
smallest_id = data_mammal['combined_mass'].idxmin()
smallest_mass = data_mammal['combined_mass'].min()
smallest_genus = data_mammal['genus'][smallest_id]
smallest_species = data_mammal['species'][smallest_id]
print(f"Smallest species: {smallest_species}, in genus: {smallest_genus}, with mass: {smallest_mass} g")

# 4.5 
print("\n4.5")

# plot a bar graph to see the average size for each group, plot the mass using kilograms for readability
sns.barplot(data_mammal, x="status", y="combined_mass_kg", hue='status')
plt.title("Average size of extant vs. extinct species")
plt.xlabel("Status")
plt.ylabel("Body mass (kg)")
plt.show()
````

Looking at the average mass of extinct and extant species overall is interesting, but there are lots of different processes that could cause size-biased extinctions so its not as informative as we might like. However, if we see the exact same pattern on each of the different continents that might really tell us that there was some worldwide event. Repeat the analysis of before, but this time compare the mean masses within each of the different continents. 

The continent labels in the data are not very consistent in the data. For example `AUS`, `OC`, and `Oceanic` can all be merged. Try to find logical combinations, if you cannot, create a seperate category. It is important to clean this up for this assignment. Again, you can use `replace()` for this.

Try to make the code as clean as possible! There is no need to copy and paste the same code for each of the categories!

### Exercises

- 4.6 Create one barplot per continent where you show the average mass of the extant and extinct species on that continent. Also print the difference between the average extant and average extinct masses for each of the continents AF, AUS, EA and SA.

````python
# check what continents there are
continents = data_mammal['continent'].unique()

# replace continent labels for consistency
data_mammal['continent'] = data_mammal['continent'].replace({
    'Oceanic': 'Oceania',
    'AUS': 'Oceania',
    'Insular': 'Americas', # Insular = somewhere in the North or South America continent or the islands. Exact location unknown
    'Af': 'AF',
})

# plot a bar graph the average mass of the extant and extinct species on each continent
sns.catplot(data=data_mammal, kind="bar", x="continent", y="combined_mass_kg", hue="status")
plt.title("Average mass of the extant and extinct species per continent")
plt.xlabel("Continent")
plt.ylabel("Body mass (kg)")
plt.show()

# group extant and extinct species by continent and take the mean of mass for each
grouped_extant = data_mammal[extant].groupby('continent')
avg_extant = grouped_extant[['combined_mass_kg']].mean()
grouped_extinct = data_mammal[extinct].groupby('continent')
avg_extinct = grouped_extinct[['combined_mass_kg']].mean()

# filter out the required continents by indexing the rows by their names
target_continents = ['EA', 'SA', 'AF', 'Oceania']
filtered_extant = avg_extant.loc[target_continents]
filtered_extinct = avg_extinct.loc[target_continents]

# calculate the difference in averages for all the required continents
difference = filtered_extinct['combined_mass_kg'] - filtered_extant['combined_mass_kg']
print('Difference between the average extant and average extinct masses per continent')
display(difference)
````

We have previously compared the average masses of extant and extinct species on different continents to try to understand whether size has an influence on extinction in mammals. Looking at the averages was a good start, but we really need to look at the full distributions of masses of the two groups to get the best picture of whether or not there was a major size bias in extinctions during the late Pleistocene.

````python
# Explore what the data for only edible mushrooms looks like
edibles = df_mushroom[df_mushroom['class'] == 'edible']
# display(edibles)

# First will check visually the bar plots comparing edible and poisononus mushrooms to see which qualities
# are usually only edible. I will be focusing on bars that show zero poisonous mushrooms to ensure that 
# the final conclusion does not include poisonous mushrooms. 

# The following plots are the ones that show a feature with a large number of edible mushrooms (range ~200-550)

# habitat: waste
sns.countplot(data=df_mushroom, x="habitat", hue="class")
plt.show()

# population: numerous, abundant
sns.countplot(data=df_mushroom, x="population", hue="class")
plt.show()

# stalk-color-below-ring: gray
sns.countplot(data=df_mushroom, x="stalk-color-below-ring", hue="class")
plt.show()

# stalk-color-above-ring: gray
sns.countplot(data=df_mushroom, x="stalk-color-above-ring", hue="class")
plt.show() 

#stalk-root: rooted
sns.countplot(data=df_mushroom, x="stalk-root", hue="class")
plt.show()

#odor: almond, anise, MAYBE none 
sns.countplot(data=df_mushroom, x="odor", hue="class")
plt.show()

#  feature where odor: none, has some poisonous mushrooms but has overwhelmingly more edible ones)
n_odor = len(df_mushroom[df_mushroom['odor'] == 'none'])
n_odor_edibles = len(edibles[edibles['odor'] == 'none'])
p_edible_odor = n_odor_edibles / n_odor
print('proportion of edible odor mushrooms: ', p_edible_odor)

# Stalk color above the ring and stalk color below the ring have the highest number of edibles compared to poisonous mushrooms.
# They both have 576 edible mushrooms when the color is gray specifically. I will focus on checking the mushrooms with these
# qualities on whether they are edible. The value matching also suggests that they might have a connection.
display(f'Number of edibles with gray stalk color above the ring: ',len(edibles[edibles['stalk-color-above-ring'] == 'gray']))
display(f'Number of edibles with gray stalk color below the ring: ',len(edibles[edibles['stalk-color-below-ring'] == 'gray']))


# Next will focus on the next feature with a high number of only edibles compared to poisonous and see 
# what it shows when combined with the previous cecked edible features. The feature is population and odor has similar numbers.
def check_condition(df, condition, label):
    match = df[condition]
    edibles = match[match['class'] == 'edible']
    poisonous = match[match['class'] == 'poisonous']
    
    print(f"\n{label}")
    print(f"total: {len(match)}")   # Total shows how many matches yields the combined condition
    print(f"edibles: {len(edibles)}")   # Edibles shows how manyof those matches are edible
    print(f"poisonous: {len(poisonous)}")  # Poisonous shows ehther the combination yields poisonous mushrooms


# 1. gray stalk above + below ring
gray_stalk = (
    (df_mushroom['stalk-color-above-ring'] == 'gray') &
    (df_mushroom['stalk-color-below-ring'] == 'gray')
)
check_condition(df_mushroom, gray_stalk, "1. gray stalk above + below ring")

# 2. Gray stalk + almond or anis odor
# We get zero matches for edibles which means that the combination doesn't work.
gray_odor = gray_stalk & df_mushroom['odor'].isin(['almond', 'anis'])
check_condition(df_mushroom, gray_odor, "2. gray stalk + almond & anis odor")

# 3. Almond or anis odor + numerous or abundant population,, 
odor_population = (
    df_mushroom['odor'].isin(['almond', 'anis']) &
    df_mushroom['population'].isin(['numerous', 'abundant'])
)
# If we combine odors with the population features, we get some matches but not as many as for  gray stalk combination
check_condition(df_mushroom, odor_population, "3. almond & anis odor + numerous & abundant population")

# 4. Gray stalk + almond or anis odor
pop_habitat = df_mushroom['habitat'].isin(['waste']) & df_mushroom['population'].isin(['numerous', 'abundant'])
check_condition(df_mushroom, pop_habitat, " 4. almond & anis odor + habitat waste")

# As gray stalk above ring + gray stalk below ring seems to yield the highest number of edible mushrooms
# will explore visually which features appear most with them
# display(df_mushroom[gray_stalk].head(50))
        
# Check the count of unique values in each column and how often they appear with the gray stalk combination
for column in df_mushroom[gray_stalk].columns:
    print(f"\nCounts for '{column}':")
    print(df_mushroom[gray_stalk][column].value_counts())
    
# Based on the visual inspection of the data, plots and the counts of values occurring with 
# the combined condition of highest number of edible mushrooms,
# I would recommend a mushroom with these features:

# bruises: yes
# odor: none
# gill-attachment: free
# gill-spacing: close
# gill-size: broad
# stalk-shape: tapering
# stalk-root: bulbous
# stalk-surface-above-ring: smooth
# stalk-surface-below-ring: smooth
# stalk-color-above-ring: gray
# stalk-color-below-ring: gray
# veil-type: partial
# veil-color: white
# ring-number: one
# ring-type: pendant
# habitat: woods

# With this combination, you get 192 different mushrooms that fit the description and are considered edible.
````
